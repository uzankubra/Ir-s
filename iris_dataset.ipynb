```python

import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

url = "pca_iris.data"
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
```

```python
df
```

```python
ozellikler = ['sepal length', 'sepal width', 'petal length', 'petal width']

x = df[ozellikler]
y = df[['target']]
```

```python
x = StandardScaler().fit_transform(x)
```

```python
x
```

```python
pca = PCA(n_components=2)
principalCN = pca.fit_transform(x)
Df = pd.DataFrame(data = principalCN, columns = ['ozellik 1', 'ozellik 2'])
```

```python
Df
```

```python
final_df = pd.concat([Df, df[['target']]], axis = 1)
```

```python
final_df.head()
```

```python
iris_setosa= final_df[df.target=='Iris-setosa']
iris_virginica = final_df[df.target=='Iris-virginica']
iris_versicolor = final_df[df.target=='Iris-versicolor']
plt.xlabel('ozellik 1')
plt.ylabel('ozellik 2')

plt.scatter(iris_setosa['ozellik 1'], iris_setosa['ozellik 2'],color='green')
plt.scatter(iris_virginica['ozellik 1'], iris_virginica['ozellik 2'],color='red')
plt.scatter(iris_versicolor['ozellik 1'], iris_versicolor['ozellik 2'],color='blue')
```
